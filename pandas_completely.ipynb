## handle Null Values

# dropna ad fill na

```python
a = pd.read_csv(r"C:\Users\LENOVO\OneDrive\Desktop\all_stocks_5yr.csv")
df = pd.DataFrame(a)
df
```

drop whole the null data 

```python
df.dropna()
```

axis = 1 kya karta hai ki bo pure data se column ko delete kr deta hai jisnme nul value store hoti hai 

```python
df.dropna(axis=1)
```

how  = "all" ka mtlb hai ki bo pure dataset se uss row ko delete karega jisme pure row mein null value store hoti hai 

```python
df.dropna(how="all")
```

ess cell mein bo sari null values remove hongi jo low ke colmn mein store hai 

```python
df.dropna(subset=["low"])
```

esme pure dataset se null value remove hoke nay dat set mein store hota hai 

```python
df.dropna(inplace= True)
```

ess cell mein thresh ka mtlb hai ki bo jitne thresh ki value doge 1,2,3,4,5,6,......... bo column ke andar check karega null values ko delete karega uss rows ko jaha null value thresh ki value ke brabar ho

```python
df.dropna(thresh=1)
```

# fill na

ye sbhi null values ko 0 se fill karega 

```python
df.fillna(0)
```

Fill using dic open fill hoga 0 se aur close 999 se

```python
df.fillna({'open':0, 'close':999})
```

forward fill aur backward fill 

```python
df.fillna(method='ffill')
```

```python
df.fillna('bfill')
```

```python
df.fillna(12,inplace=True)
df
```
