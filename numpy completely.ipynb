# Datatype in Numpy Arrays

Check data type

```python
var = np.array([5,6,78,9])
print(var.dtype)
print("\n")
var1 = np.array([5,6,78,9])
print(var1.dtype)
print("\n")
var2 = np.array([5,6,78,9,"a","b","c",7,8,9])
print(var2.dtype)
```

Store data from int64 to int8

```python
x = np.array([5,6,7,8], dtype = np.int8)
print(x.dtype)
```

Convert in to another after changing the dtype value 

```python
x = np.array([5,6,7,8], dtype = "f")
print(x.dtype)
```

convert the data type of an array in to another datatype using dtype function 

```python
x = np.array([5,6,7,8])
new = np.float32(x)
new_1 = np.int8(new)
print(x.dtype)
print(new.dtype)
print(new_1.dtype)
```

Directly change the datatype 

Astype function is used to convert the data type of an array into another data type.

```python
new_1 = new.astype(float)
print(new_1)
print(new_1.dtype)
```

# Arithmetic Functions

```python
A = np.array([1,4,5,6,73,5,6,7,7,8,8])
print("max",np.max(A))
print("min",np.min(A))   
```

```python
print("maximmum with index value.:",np.max(A),np.argmax(A))
print("minimmum with index value.:",np.min(A),np.argmin(A))
```

```python
B = np.array([[1,2,3,4],[5,6,78,7]])
print(B)
print("max:", np.max(B, axis = 0))  # column wise maximum
print("max:", np.max(B, axis = 1))  # row wise maximum
```
