# Join and Split the array

## join
joining means putting content of two or more arrays in a single array

```python
j = np.array([1,2,3,4,5])
k = np.array([6,7,8,9,10])
ar = np.concatenate((j,k))
print(ar)
```

```python
l = np.array([[1,2,3,4,5],[34,5,6,77,44]])
m = np.array([[6,7,8,9,10],[22,3,44,52,6]])
ar1 = np.concatenate((l,m))
print(ar1)
```

```python
l = np.array([[1,2,3,4,5],[34,5,6,77,44]])
m = np.array([[6,7,8,9,10],[22,3,44,52,6]])
ar1 = np.concatenate((l,m),axis = 1)
print(ar1)
```

```python
l = np.array([[1,2,3,4,5],[34,5,6,77,44]])
m = np.array([[6,7,8,9,10],[22,3,44,52,6]])
ar3 = np.concatenate((l,m),axis = 0)
print(ar3)
```

```python
j = np.array([1,2,3,4,5])
k = np.array([6,7,8,9,10])
ar = np.stack((j,k),axis=1)
print(ar)
```

```python
j = np.array([1,2,3,4,5])
k = np.array([6,7,8,9,10])
ar = np.hstack((j,k))# rows
ar1 = np.vstack((j,k))# columns
ar2 = np.dstack((j,k))# height
print(ar)
print(ar1)
print(ar2)
```
